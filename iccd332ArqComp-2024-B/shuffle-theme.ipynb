# Shuffle Theme

```python
import random
import pandas as pd

# Number of students
num_students = 23  # Adjust as needed

# List of project themes
themes = ["IEEE 754 Conversion", "IEEE Float Sum", "IEEE Float Product"]

# Generate a list of students as numbers
students = list(range(1, num_students + 1))

# Shuffle students randomly
random.shuffle(students)
```

```python
# Specify solo students manually (as a single number or a list)
solo_students = [14]  # Adjust as needed

# Remove solo students from the main list
students = [s for s in students if s not in solo_students]

# Ensure an even number of students before pairing
if len(students) % 2 != 0:
    extra_solo = students.pop()  # Remove the last student if list is still odd
    solo_students.append(extra_solo)
```

```python
# Pair the remaining students and assign a random project theme
pairs = [(students[i], students[i+1], random.choice(themes)) for i in range(0, len(students), 2)]

# Create pandas DataFrame for pairs
df_pairs = pd.DataFrame(pairs, columns=["Student 1", "Student 2", "Project Theme"])

# Assign random themes to solo students
df_solo = pd.DataFrame(
    [[solo, "-", random.choice(themes)] for solo in solo_students], 
    columns=["Student 1", "Student 2", "Project Theme"]
)
```

```python
# Combine both DataFrames
df_final = pd.concat([df_pairs, df_solo], ignore_index=True)

# Print DataFrame
print("Assigned Projects:")
df_final
```

```python
from string import ascii_uppercase

# Read Excel file into a DataFrame
excel_df = pd.read_excel('D:\\entea\\Lectures\\iccd332ArqComp-2024-B\\StudentsList.xlsx', sheet_name="ListaEstudiantes2025-A")  # Replace with your actual file path

# Create a mapping from student number to group letter and theme
group_map = {}
theme_map = {}

column_idx = "Nro."
for idx, row in df_final.iterrows():
    group_letter = ascii_uppercase[idx]
    s1 = row['Student 1']
    s2 = row['Student 2']
    theme = row['Project Theme']
    group_map[s1] = group_letter
    theme_map[s1] = theme
    if s2 != '-':
        group_map[int(s2)] = group_letter
        theme_map[int(s2)] = theme

# Add 'Group' and 'Theme' columns to the Excel DataFrame based on 'Nro'
excel_df['Group'] = excel_df[column_idx].map(group_map)
excel_df['Theme'] = excel_df[column_idx].map(theme_map)

excel_df
```

```python
excel_df.to_excel('D:\\entea\\Lectures\\iccd332ArqComp-2024-B\\AssignedProjects2025-A.xlsx', index=False)  # Save to a new Excel file
print("Projects assigned and saved to 'AssignedProjects2025-A.xlsx'.")
```
