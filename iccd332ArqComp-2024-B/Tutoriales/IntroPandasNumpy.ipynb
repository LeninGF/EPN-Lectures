## Numpy

Haremos una breve introducción a operaciones fundamentales con numpy. En este caso se declaran, un vector unidimensional, una matriz de dos dimensiones y una de tres dimensiones con valores aleatorios.

```python
import numpy as np
np.random.seed(45)
x1 = np.random.randint(0, 10, size=5)
print(x1)
print(x1.shape)
x2 = np.random.randint(10, size=(3,4))
print(x2)
print(x2.shape)
x3 = np.random.randint(10, size=(3,4,5))
print(x3)
print(x3.shape)
```

Para acceder a valores particulares de un vector o matriz se utiliza la indexación

```python
x3[1,2,3]
```

Para cortar o truncar un dato numpy se ha de seguir el formato `x[start:stop:step]`

```python
x = np.linspace(-3*np.pi, 3*np.pi, 1000)
y = np.sin(x)/x
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.title("Sinc Function")
plt.xlabel("x")
plt.ylabel("sinc(x)")
plt.grid()
plt.show()
```

Se obseva que los tipos de dato no han sido detectados de la mejor manera. De ahi que sea conveniente hacer conversiones para adecuarlos o leer nuevamente el dataset directamente con las modificaciones pertinentes si estas se conocen a priori

```python
df['NDD_OFUSCADA'] = df['NDD_OFUSCADA'].astype('str')
df['NDD_OFUSCADA'] = df['NDD_OFUSCADA'].str.zfill(10)
```

```python
# df.FECHA_INCIDENTE = pd.to_datetime(df.FECHA_INCIDENTE, format='%Y-%m-%d')
# La anterior producira un error
df.FECHA_INCIDENTE = pd.to_datetime(df.FECHA_INCIDENTE, format='%Y-%m-%d', errors='coerce')
# Con el formato correcto, extraemos el año
df['AÑO_INCIDENTE'] = df.FECHA_INCIDENTE.dt.year
df['MES_INCIDENTE'] = df.FECHA_INCIDENTE.dt.month
```
