## tabla cruzada

Se puede obtener una tabla que obtenga la cuenta de desagregaciones del delito de robo por año

```python
pd.crosstab(df_quito.MODALIDAD_DESAGREGACION, df_quito.AÑO_INCIDENTE, margins=True, margins_name="TOTAL")#.sort_values(by='TOTAL', ascending=False)
```

Si la tabla anterior se normaliza y se combina con un expectro de colores se puede obtener un mapa de calor

```python
import seaborn as sns
tabla = pd.crosstab(
    df_quito.MODALIDAD_DESAGREGACION,
    df_quito.AÑO_INCIDENTE
)

plt.figure(figsize=(10,6))
sns.heatmap(tabla, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Mapa de calor de incidentes por modalidad y año")
plt.show()
```

Se puede normalizar la tabla segun el total de cada fila o el total por columnas

```python
tabla_norm = tabla.div(tabla.sum(axis=1), axis=0)
sns.heatmap(tabla_norm, annot=True, cmap="YlGnBu", fmt=".2f")
```

Se puede normalizar con respecto a la columna (i.e. por año)


```python
tabla_norm = tabla.div(tabla.sum(axis=0), axis=1)
sns.heatmap(tabla_norm, annot=True, cmap="YlOrRd", fmt=".2f")
```

Se puede normalizar con respecto al total de la tabla cruzada

```python
tabla_norm = (tabla - tabla.min().min()) / (tabla.max().max() - tabla.min().min())
sns.heatmap(tabla_norm, cmap="coolwarm")
```
