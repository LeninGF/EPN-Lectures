Generación de Equipos de trabajos para presentación de temas

```python
import pandas as pd
lista_path = 'listaEstudiantes\ArquitecturaComputadoresListaCalificaciones2024Ahhdd.xlsx'
sheet_name = "ArqComp"
student_df = pd.read_excel(lista_path, sheet_name=sheet_name)
student_df.shape
```

```python
student_df_new = student_df.copy()
student_df_new = student_df[-student_df["Prueba"].isna()]
```

```python
student_df_new.shape
```

```python
import numpy as np
result_dict = {}
team_labels = ['E1', 'Q1', 'E2','Q2', 'E3', 'Q3', 'E4', 'Q4']
weeks = ['W1', 'W2', 'W3', 'W4', 'W5']
for r, w in enumerate(weeks):
    df = student_df_new.sample(frac=1).reset_index(drop=True)
    df[w] = ""
    num_teams = len(team_labels)
    for i, row in enumerate(df.iterrows()):
        team_idx = i%num_teams
        df.at[row[0], w] = team_labels[team_idx]
    result_dict[w] = df    
```

```python
lista_dataframes  = []
for key in result_dict.keys():
    this_dict = dict(zip(result_dict[key]['Código'], result_dict[key][key]))
    this_df = pd.DataFrame(this_dict.items(), columns=['Código', key])
    lista_dataframes.append(this_df)
```

```python
result_df = student_df_new.copy()
for other_df in lista_dataframes:
    result_df = pd.merge(result_df, other_df, on="Código", how='inner')
result_df.shape
```

```python
temp_df = pd.read_excel('listaEstudiantes/EstudiantsShuff.xlsx')
temp_df.shape
```

```python
def total_expo(*args):
    cont_expo = 0
    for col in args:
        if 'E' in  col:
            cont_expo+=1
    return cont_expo

def total_quest(*args):
    cont_quest = 0
    for col in args:
        if 'Q' in  col:
            cont_quest+=1
    return cont_quest

temp_df['Total_E'] = temp_df.apply(lambda x: total_expo(x['W1'], x['W2'], x['W3'], x['W4'], x['W5']), axis=1)
temp_df['Total_Q'] = temp_df.apply(lambda x: total_quest(x['W1'], x['W2'], x['W3'], x['W4'], x['W5']), axis=1)
```
